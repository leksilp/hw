```python
import matplotlib.pyplot as plt
import pandas as pd
from pylab import rcParams

rcParams['figure.figsize'] = 13,6

import seaborn as sns
```

```python
data19 = pd.read_csv('2019.csv')

data19['Year'] = 2019
data19

data18 = pd.read_csv('2018.csv')
data18['Year'] = 2018
data18

rating = pd.concat([data19,data18])
rating.head()
```

```python
rating_year = rating.pivot_table(index ='Country or region', columns = 'Year', values = 'Healthy life expectancy', 
                    fill_value = 0)
rating_year.sort_values(by=2019, ascending=False).head(8).plot()
plt.title('Level healthy life expectancy')
plt.ylabel('Score')

rating_year.sort_values(by=2019, ascending=False).tail(8).plot()
plt.title('Level healthy life expectancy')
plt.ylabel('Score')
```

```python
rating_year = rating.pivot_table(index ='Country or region', columns = 'Year', values = 'Social support', 
                    fill_value = 0)
rating_year.sort_values(by = 2018, ascending=False).tail(8).plot(kind='area',
                             stacked=False)
plt.title('Level of social support')
plt.ylabel('Social support')
```

```python
rating_year = rating.pivot_table(index ='Country or region', columns = 'Year', values = 'Score', 
                    fill_value = 0)
rating_year.sort_values(by = 2018, ascending=False).head(15).plot(kind='bar')
plt.title('Level of happiness')
plt.ylabel('Score')

rating_year.sort_values(by = 2018, ascending=False).tail(15).plot(kind='bar')
plt.title('Level of happiness')
plt.ylabel('Score')
```
