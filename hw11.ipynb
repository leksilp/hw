```python
from pymystem3 import Mystem
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

data
```

```python
m = Mystem()
```

```python
def lemm(word):
    return(' '.join(m.lemmatize(word)))
```

```python
data['lemmas'] = data['keyword'].apply(lemm)
data
```
